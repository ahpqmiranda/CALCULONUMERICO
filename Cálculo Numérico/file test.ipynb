# ELABORAÇÃO DE GRÁFICOS COM A BIBLIOTECA MATPLOTLIB E NUMPY


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x): return np.exp(x)-x+2
```

```python
xx = np.linspace(-5,5)
plt.plot(xx,f(xx))
plt.grid(); plt.show()
```

```python
def fl(x): return np.exp(x)-1
```

```python
plt.plot(xx,fl(xx))
plt.grid();plt.show()
```
